### Hypothetical Statement - 1

#### 1. State Your research hypothesis as a null hypothesis and alternate hypothesis.

Null Hypothesis (H0): There is no significant relationship between Quantity purchased and Revenue generated.

Alternative Hypothesis (H1): There is a significant relationship between Quantity purchased and Revenue generated.

````python
from scipy.stats import pearsonr

corr, p_value = pearsonr(df['Quantity'], df['Revenue'])

print("Correlation:", corr)
print("P-value:", p_value)
````

##### Which statistical test have you done to obtain P-Value?

### Hypothetical Statement - 2

#### 1. State Your research hypothesis as a null hypothesis and alternate hypothesis.

Null Hypothesis (H0): The average Unit Price is the same across all countries.

Alternative Hypothesis (H1): The average Unit Price differs significantly across countries.

#### 2. Perform an appropriate statistical test.

````python
from scipy.stats import f_oneway

country_groups = [
    group['UnitPrice']
    for name, group in df.groupby('Country')
    if len(group) > 100
]

f_stat, p_value = f_oneway(*country_groups)

print("F-statistic:", f_stat)
print("P-value:", p_value)
````

##### Which statistical test have you done to obtain P-Value?

ANOVA (Analysis of Variance) Test was used to compare the average Unit Price across multiple countries and determine whether significant differences exist.

##### Why did you choose the specific statistical test?

ANOVA is appropriate because it compares the means of a numerical variable (UnitPrice) across more than two groups (Countries). It helps determine whether pricing differs significantly among countries.

f p-value < 0.05:

The null hypothesis is rejected. There is a statistically significant difference in average Unit Prices across countries.

If p-value ≥ 0.05:

The null hypothesis is accepted. There is no statistically significant difference in average Unit Prices across countries.


### Hypothetical Statement - 3

#### 1. State Your research hypothesis as a null hypothesis and alternate hypothesis.

Null Hypothesis (H0): High-spending customers and low-spending customers have the same average purchase quantity.

Alternative Hypothesis (H1): High-spending customers and low-spending customers have different average purchase quantities.

#### 2. Perform an appropriate statistical test.

````python
from scipy.stats import ttest_ind

median_revenue = df['Revenue'].median()

high = df[df['Revenue'] > median_revenue]['Quantity']
low = df[df['Revenue'] <= median_revenue]['Quantity']

t_stat, p_value = ttest_ind(high, low)

print("T-statistic:", t_stat)
print("P-value:", p_value)
````

##### Which statistical test have you done to obtain P-Value?

Independent T-Test was used to compare the average purchase quantity between high-spending customers and low-spending customers.

##### Why did you choose the specific statistical test?

The Independent T-Test is suitable because it compares the means of a numerical variable (Quantity) between two independent groups (high-spending and low-spending customers). It helps determine whether the difference in purchase quantity between the two groups is statistically significant.

If p-value < 0.05:

The null hypothesis is rejected. There is a statistically significant difference in purchase quantity between high-spending and low-spending customers.

If p-value ≥ 0.05:

The null hypothesis is accepted. There is no statistically significant difference in purchase quantity between high-spending and low-spending customers.

### 2. Handling Outliers

````python
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'])

snapshot_date = df['InvoiceDate'].max() + pd.Timedelta(days=1)

rfm = df.groupby('CustomerID').agg({
    'InvoiceDate': lambda x: (snapshot_date - x.max()).days,
    'InvoiceNo': 'nunique',
    'Revenue': 'sum'
}).reset_index()

rfm.columns = ['CustomerID', 'Recency', 'Frequency', 'Monetary']
````

````python
for col in ['Recency', 'Frequency', 'Monetary']:
    Q1 = rfm[col].quantile(0.25)
    Q3 = rfm[col].quantile(0.75)
    IQR = Q3 - Q1

    outliers = ((rfm[col] < Q1 - 1.5*IQR) |
                (rfm[col] > Q3 + 1.5*IQR)).sum()

    print(f"{col}: {outliers} outliers")
````

##### What all outlier treatment techniques have you used and why did you use those techniques?

The Interquartile Range (IQR) method was used to identify outliers in the Recency, Frequency, and Monetary features. The analysis detected 124 outliers in Frequency and 122 outliers in Monetary, while no outliers were found in Recency. These outliers were retained because they represent genuine customer behavior, such as loyal customers and high-value customers, which are important for RFM-based customer segmentation. Retaining them helps preserve meaningful business insights and improves the effectiveness of clustering.

### ML Model - 1

````python
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters=2, random_state=42, n_init=10)

rfm['Cluster_KMeans'] = kmeans.fit_predict(rfm_scaled)

print(rfm.head())
````

````python
import matplotlib.pyplot as plt

score = 0.36

plt.figure(figsize=(5,4))
plt.bar(['Silhouette Score'], [score])
plt.title('K-Means Evaluation Metric')
plt.ylabel('Score')
plt.ylim(0,1)
plt.show()
````

K-Means Clustering was used to segment customers based on their RFM (Recency, Frequency, Monetary) values. The algorithm grouped customers into 4 clusters with similar purchasing behavior. Model performance was evaluated using the Silhouette Score, which measures how well data points fit within their assigned clusters compared to other clusters. The obtained Silhouette Score was 0.616, indicating good cluster separation and meaningful customer segmentation. This model can help businesses identify high-value customers, regular customers, and at-risk customers for targeted marketing strategies.

#### 2. Cross- Validation & Hyperparameter Tuning

````python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

scores = []

for k in range(2, 8):
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    labels = kmeans.fit_predict(rfm_scaled)

    score = silhouette_score(rfm_scaled, labels)
    scores.append((k, score))

for k, score in scores:
    print(f"K = {k}, Silhouette Score = {score:.3f}")
````

````python
import matplotlib.pyplot as plt

k_values = [k for k, score in scores]
sil_scores = [score for k, score in scores]

plt.figure(figsize=(6,4))
plt.plot(k_values, sil_scores, marker='o')
plt.xlabel("Number of Clusters (K)")
plt.ylabel("Silhouette Score")
plt.title("Hyperparameter Tuning using Silhouette Score")
plt.grid(True)
plt.show()
````

##### Which hyperparameter optimization technique have you used and why?

Hyperparameter tuning was performed by evaluating K-Means clustering with different numbers of clusters (K=2 to K=7). The Silhouette Score was used as the evaluation metric. The highest Silhouette Score of 0.896 was obtained when K=2, indicating the best cluster separation and cohesion. Therefore, K=2 was selected as the optimal number of clusters for the final K-Means model.

##### Have you seen any improvement? Note down the improvement with updates Evaluation metric Score Chart.

After hyperparameter tuning, the K-Means model achieved a Silhouette Score of 0.896, which is higher than the scores obtained for other cluster values and other clustering models. This demonstrates improved clustering performance and more meaningful customer segmentation.

#### 2. Cross- Validation & Hyperparameter Tuning

````python
from sklearn.cluster import AgglomerativeClustering
from sklearn.metrics import silhouette_score
import matplotlib.pyplot as plt

scores = []

for k in range(2, 8):
    hc = AgglomerativeClustering(n_clusters=k)
    labels = hc.fit_predict(rfm_scaled)
    score = silhouette_score(rfm_scaled, labels)
    scores.append(score)

best_k = range(2, 8)[scores.index(max(scores))]

print("Best Number of Clusters:", best_k)
print("Best Silhouette Score:", round(max(scores), 3))

plt.figure(figsize=(6,4))
plt.plot(range(2, 8), scores, marker='o')
plt.xlabel("Number of Clusters")
plt.ylabel("Silhouette Score")
plt.title("Hyperparameter Tuning - Hierarchical Clustering")
plt.show()
````

Silhouette Score-based hyperparameter tuning was used for Hierarchical Clustering.
